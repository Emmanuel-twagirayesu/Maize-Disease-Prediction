```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix 
import matplotlib.pyplot as plt
import pickle
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('maize_disease_weather.csv')
DF=df.dropna()
DF.tail()
```

```python
X=DF[['Rainfall_mm','Temperature_C','Humidity_%']]
Y=DF['Disease']
y=set(Y)
print(y)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2, random_state=0)
```

```python
Model=RandomForestClassifier(n_estimators=100,random_state=0)
```

```python
Model.fit(X_train,Y_train)
```

```python
Input=[[5.45,17.89,50.67]]
Predict=Model.predict(X_test)
# Predict
```

```python
plt.figure(figsize=(5, 4))
sns.scatterplot(data=DF, x='Rainfall_mm', y='Temperature_C', hue='Disease', palette='Set1')
plt.title('Rainfall vs. Temperature by Disease')
# plt.savefig('rainfall_vs_temperature.png')
# plt.close()
```

```python
sns.pairplot(DF, vars=['Rainfall_mm', 'Temperature_C', 'Humidity_%'], hue='Disease', palette='Set1', diag_kind='hist')
plt.suptitle('Pair Plots of Features by Disease', y=1.02)
# plt.savefig('pair_plots.png')
# plt.close()
```

```python
Accuracy=accuracy_score(Y_test,Predict)
Report=classification_report(Y_test,Predict)
Matrix=confusion_matrix(Y_test,Predict)
```

```python
print("Accuracy=",Accuracy)
print("Report=\n",Report)
print("Confusion Matrix=\n",Matrix)
```
